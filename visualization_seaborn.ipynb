```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mtcars=pd.read_csv('Cars.csv')
```

```python
mtcars.head()
```

```python
mtcars.shape
```

```python
mtcars.columns
```

```python
pd.crosstab(mtcars.HP,mtcars.MPG).head(5)
```

```python
pd.crosstab(mtcars.HP,mtcars.VOL).plot(kind='bar')
```

```python
mtcars['VOL'].value_counts()
```

```python
mtcars['VOL'].value_counts().plot(kind='pie')
```

```python
plt.scatter(mtcars.VOL,mtcars.MPG)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('scatter plot')
```

```python
plt.hist(mtcars['VOL'])
plt.title('HISTOGRAM')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.hist(mtcars['VOL'],facecolor='green',edgecolor='blue',bins=5)
plt.title('Histogram')
```
