```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python

tips=sns.load_dataset('tips')
```

```python
tips
```

```python
tips.shape
```

## ***STRIP*** ***PLOT***

```python
sns.stripplot(y='tip',data=tips,jitter=True)
```

```python
sns.stripplot(y='tip',data=tips,jitter=False)
```

```python
sns.stripplot(x='day',y='tip',data=tips)
plt.ylabel('tips($)')
plt.show()
```

## SWARM PLOT

```python
sns.swarmplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='sex')
plt.ylabel('tip($)')
plt.show()
```

```python
sns.swarmplot(x='tip',y='day',hue='sex', data=tips,orient='h')
plt.xlabel('tips($)')
plt.show()
```

BOXPLOT

```python
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
```

```python
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

## COMBINING PLOTS

```python
sns.violinplot(x='day',y='tip',data=tips, inner=None, color='lightgray')
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
plt.ylabel('tip($)')

plt.show()
```

# TWO VARIABLE AND MORE

JOINT **PLOT**

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

DENSITY **PLOT**

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
plt.show()
```

# PAIR PLOT

```python
sns.pairplot(data=tips)
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```
