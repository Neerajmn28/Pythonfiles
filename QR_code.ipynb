```python
#import libraries
import pyqrcode
import matplotlib.pyplot as plt

number = 7306562096
qr_code = pyqrcode.create(number)
qr_code.png('tmb.png', scale=8) #This parameter specifies the scale of the qr_code

img = plt.imread('tmb.png') #This is the filename to which the PNG image of the QR code will be saved.
plt.imshow(img)
plt.axis('off')
plt.show()
```
